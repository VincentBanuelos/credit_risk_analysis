```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

import scipy.stats as stats

pd.options.display.max_columns = None

df = pd.read_csv('loan_data_2007_2014.csv')
df.head(10)
```

```python
df.drop(['Unnamed: 0','id', 'url', 'desc'], axis=1, inplace=True)
null_counts = df.isnull().sum()
df.columns.to_list()
```

```python
df.emp_length.fillna(value=str(0), inplace=True)

df['emp_length'] = df.emp_length.str.replace('+ years','')
df['emp_length'] = df.emp_length.str.replace('years','')
df['emp_length'] = df.emp_length.str.replace('year','')
df['emp_length'] = df.emp_length.str.replace('< 1','0')
df['emp_length'] = df['emp_length'].str.strip()
df['emp_length'] = pd.to_numeric(df['emp_length'])
df.emp_length.value_counts()
```

```python
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], format='%b-%y')
type(df['earliest_cr_line'][0])
```

```python
df['days_since_earliest_cr_line'] = (pd.to_datetime('12-1-2017') - df['earliest_cr_line'])
df['days_since_earliest_cr_line']
```

```python
df['months_since_earliest_cr_line'] = round(pd.to_numeric((df['days_since_earliest_cr_line'] / np.timedelta64(1, 'D')) / 30))
df['months_since_earliest_cr_line']
```

```python
df['months_since_earliest_cr_line'].describe()
```

```python
negative_values = df.loc[df['months_since_earliest_cr_line'] < 0, ['months_since_earliest_cr_line', 'earliest_cr_line']]
negative_values
```
